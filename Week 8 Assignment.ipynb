### Import pandas as pd, numpy as np, seaborn as sns.

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

### Import and read the data. 

```python
auto_mpg = pd.read_table('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data', 
                   sep = '\s+', 
                   names =['MPG', 'Cylinders', 'Displacement', 'Horsepower', 
                           'Weight', 'Acceleration', 'Model Year', 'Origin','Car Name'], 
                   header = None).set_index('Car Name')
auto_mpg.head(35)
```

### Some of the horsepower data is missing and marked with a ‘?’. Therefore, all the '?' are replaced with with a pandas-appropriate missing value (NaN).

```python
auto_mpg.replace('?', np.nan, inplace=True)

auto_mpg.head(35)
```

### Convert the "Horsepower" column to numeric as well.

```python
auto_mpg.dtypes
```

```python
auto_mpg['Horsepower'] = pd.to_numeric(auto_mpg['Horsepower']
                                       , errors = 'coerce')
auto_mpg.dtypes
```

###  The 'Origin' column values are 1, 2, and 3. These values are converted to ‘USA’, ‘Asia’, and ‘Europe’.

```python
auto_mpg['Origin'].replace({1:'USA',
                            2:'Asia',
                            3:'Europe'},
                           inplace = True)

auto_mpg.head(35)
```

### This bar chart shows the distribution for cylinders.

```python
sns.set_style('whitegrid')

sns.barplot(data = auto_mpg,
            x = 'Origin',
            y = 'MPG',
            hue = 'Cylinders',
            color = 'turquoise')
sns.set()
```

### This scatterplot shows the relationship between horsepower and weight.

```python
sns.lmplot(data = auto_mpg,
           x = 'Horsepower',
           y = 'Weight',
           hue = 'Origin',
           height = 6)

plt.title('Horsepower vs Weight')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
```

### What is the relationship between the displacement and horsepower of the cars? 

#### Matplotlib graphics and seaborn graphics:

```python
plt.plot([2,4,6,8],
         linestyle = '-',
        marker = '^',
        color = 'turquoise')

plt.xlabel('Horsepower')
plt.ylabel('Displacement')
plt.show()
```

```python
sns.pointplot(x = 'Displacement',
              y = 'Horsepower',
              data = auto_mpg,
             markers = 'o',
             color = 'turquoise')
```
